# 3.1. Series

3.1.8. Створіть Series зі списку рядків. Визначте довжину
кожного рядка та знайдіть найдовший.

```python
import pandas as pd

data = ["Київ", "Львів", "Вінниця", "Полтава", "Дніпро"]
series = pd.Series(data)

lengths = series.str.len()

longest_string = series[lengths == lengths.max()]

print("Початковий Series:")
print(series, "\n")

print("Довжина кожного рядка:")
print(lengths, "\n")

print(f"Найдовший рядок: \n {longest_string} ")
```

**3.5.8. Об’єднайте таблиці students і scores.**
Виберіть студентів, які склали всі предмети (score ≥ 60).
Порахуйте їх середній бал.


```python
import pandas as pd

st = pd.read_csv("students.csv")
sc = pd.read_csv("scores.csv")

result = pd.merge(st, sc, on='student_id', how='inner')

failed_students = result[result['score'] < 60]['student_id'].unique()

# (~ міняє умову на протилежну — "не входить до списку тих, хто не склав")
successful_students_df = result[~result['student_id'].isin(failed_students)]
average = successful_students_df['score'].mean()

print(f'Середній бал студентів, які склали всі предмети: {average:.2f}')
successful_students_df
```
